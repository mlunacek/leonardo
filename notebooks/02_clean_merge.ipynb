```python
import json
import glob
```

```python
filenames = glob.glob("raw/*.json")
```

```python
data = []

for filename in filenames:
    
    tmp = json.loads(open(filename).read())
    for item in tmp:
        item['index'] = int(item['index'])
        item['location'] = item['location'].replace(u'\xa0',' ')
        item['distance'] = item['distance'].replace(u'\xa0',' ')
        item['xckm'] = item['xckm'].replace(u'\xa0',' ')
        data.append(item)
    
data = sorted(data, key=lambda x: x['index'])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data).drop_duplicates()
```

```python
df.to_csv("clean/rmhpa_flights.csv", index=False)
```
